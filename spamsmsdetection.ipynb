```python
import pandas as pd
data = pd.read_csv("spam.csv",encoding ='ISO-8859-1')
data.rename(columns={"v1":"Status",
                     "v2":"Text1",
                     "Unnamed: 2":"Text2",
                     "Unnamed: 3":"Text3",
                     "Unnamed: 4":"Text4"}, inplace= True)
data
```

```python
data.isnull().sum()
```

```python
data['Text']= data['Text1'].fillna(" ").astype(str)+" "+data['Text2'].fillna(" ").astype(str)+" "+data['Text3'].fillna(" ").astype(str)+" "+data['Text4'].fillna(" ").astype(str)
data = data.drop(columns = ['Text1','Text2','Text3','Text4'])
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["Status"]=le.fit_transform(data["Status"])
data
```

```python
x = data["Text"]
y = data["Status"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2 , random_state = 42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words ='english',max_features=5000)
X_train_tfidf = vectorizer.fit_transform(x_train)
X_test_tfidf = vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)
y_pred_nb = nb_model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print("Naive Bayes Classifier : ")
print(classification_report(y_test, y_pred_nb))
print("Accuracy : ", accuracy_score(y_test, y_pred_nb))
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train_tfidf, y_train)
y_pred_lr = lr_model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print("Logistic Regression Classifier :")
print(classification_report(y_test, y_pred_lr))
print("Accuracy : ", accuracy_score(y_test, y_pred_lr))
```

```python
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(X_train_tfidf, y_train)
y_pred_svm = svm_model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import classification_report , accuracy_score
print("\nSupport Vector Machine Classifier : ")
print(classification_report(y_test, y_pred_svm))
print("Accuracy : ", accuracy_score(y_test, y_pred_svm))
```

```python
new_messages = [
    "Congratulates ! you've won a free vacation to the bahamas.call now to claim your prize !",
    "Hey, are we still on for the meeting tomorrow?",
    "you've been selected for a $1000 shopping spree ! Click here to claim your gift card !",
    "Can We meet tomorrow ?"
]

new_messages_tfidf = vectorizer.transform(new_messages)
predictions_nb = nb_model.predict(new_messages_tfidf)
predictions_lr = lr_model.predict(new_messages_tfidf)
predictions_svm = svm_model.predict(new_messages_tfidf)

label_map = {0 : 'ham',1:'spam'}
for i,message in enumerate(new_messages):
  print(f"Message : {message}")
  print(f"Status : \tNaive Bayes Prediction : {label_map[predictions_nb[i] ]}")
  print(f"\t\tLogistic Regression Prediction : {label_map[predictions_lr[i]]}")
  print(f"\t\tSVM Prediction : {label_map[predictions_svm[i] ]}")
  print("=" * 100)

```
